```python
import numpy as np
import random
import h5py
import tqdm
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
predicted = np.load('PredictedLabels_Mg22_size1024.npy') #load in predictions 
# print(predicted.shape)
# print(predicted) 
```

```python
# Prints what event should look like based on particle ids of original file
file_path = '../pointnet_data/Mg22_size1024_test.h5'
one = h5py.File(file_path, 'r') # uncomment for h5 file
# data = np.load('Mg22_Ne20pp_8MeV_allarrays_with_track_id.npy')


keys = list(one.keys())
# x = 1 #number of desired events to plot
change = [] # container for all particle ids to plot
events = [0] #insert desired event(s) to plot
fig = plt.figure(figsize = (15,15))
color = ['b','g','r','y','k','m','c'] #characters corresponding to color 
count = 1


# for n in range(x): # to plot number of events 
for n in events: # uncomment to plot certain events
    ax = fig.add_subplot(1,1, count, projection='3d') 
    evt = one[keys[n]]
    # evt = data[n][:event_lens[n]] # uncomment for npy array
    
    for i,e in enumerate(evt):
        x = e[0] #get x value of instance
        y = e[1] #get y value of instance
        z = e[2] #get y value of instance
        change = np.append(change, e[5]) # add particle id
        ax.scatter3D(x,z, y, c = color[int(e[5])], s = 2) # choosing colors based on real particle id value
        
    ax.set_xlabel('X Label')
    ax.set_ylabel('Z Label')
    ax.set_zlabel('Y Label')
    num = len(np.unique(change)) -1 # find number of tracks in event if normalized, remove -1

    if num == 0:
        plt.title(keys[n] +': Beam')
    else:
        plt.title(keys[n]  +': ' + str(num) +' tracks')
    count += 1

plt.show()
# plt.savefig('Real_events_plot.png')
```
